```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import os
```

```python
raw_data_path = '../data/clean/'
all_files = os.listdir(raw_data_path)
```

```python
col_names = ['ax','ay','az','gx','gy','gz']


fitness_data = {}
for f in all_files:
    if f == 'placeholder':
        continue
    
    print(f)
    data = pd.read_csv(raw_data_path + f, header=None)
    
    if data.shape[1]!=7:
        if data.shape[1]!=10:
            print('bad file: '+f)
            continue
        data.drop([7,8,9], axis=1, inplace=True)
    fitness_data[f.split('.')[0]] = data
    data.dropna(inplace= True)
    data.columns = names=['time','ax','ay','az','gx','gy','gz']
    print(data.head())
    
```

```python
for key, df in fitness_data.items():
    a_sum = np.sqrt(df['ax']**2 + df['ay']**2 + df['az']**2) 
    a_sum = a_sum.rolling(20).mean()
    df['a_sum'] = a_sum
    df.dropna(inplace=True)
    
    valid = (df['a_sum']<0.95) | (df['a_sum']>1.05).astype(int)
    valid = valid.rolling(50).max()
    df['valid'] = valid
    df.dropna(inplace=True)
    
```
